### Input Layer: 
Takes the shape of the data that you're passing into your network so columns or features. 

### Hidden Layer: 
Deep learning layer, mysterious and don't interact with them at all.  Hidden because we don't have control.  Process that decides   the optimal weights.   

### Output Layer: 
Key thing that differentiates between regression or classification and based on how the output layer is defined we should be able to understand what the network is outputting. binary classification: sigmoid. Multiclass classification: softmax. Regression: no activation function. 

### Neuron:
A neuron is a mathematical function that takes some inputs, weight them, and then return an output

### Weight:
The weights are some parameters used within the hidden layers to transform the input

### Activation Function:
The activation function transforms the output in a specific format for our context. For example, if we have a binary classification problem, we can set a sigmoid function as activation function to transform the output vector in the probability of having the primary class. We can use different activation functions for different problems to get the output in a format suitable for the problem we are trying to solve.

### Node Map:
A node map is a schema of the neurons, which shows the number of input layers, hidden layers, and output layers, as well as the interaction between them.

### Perceptron:
A perceptron is a simple neuron.

```python
import pandas as pd
data = { 'x1': [0,1,0,1],
         'x2': [0,0,1,1],
         'y':  [1,1,1,0]
       }

df = pd.DataFrame.from_dict(data).astype('int')
```

```python
df.head()
```

```python
inputs = df.to_numpy()
inputs
```

```python
correct_outputs = [[1], [1], [1], [0]]
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    sx = sigmoid(x)
    return sx * (1 - sx)
```

```python
import numpy as np

weights = 2 * np.random.random((3,1)) - 1
weights
```

```python
# Update our weights 10,000 times - (fingers crossed that this process reduces error)
for iteration in range(10000):
    
    # Weighted sum of inputs / weights
    weighted_sum = np.dot(inputs, weights)
    
    # Activate!
    activated_output = sigmoid(weighted_sum)
    
    # Cac error
    error = correct_outputs - activated_output
    adjustments = error * sigmoid_derivative(weighted_sum)
    
    # Update the Weights
    weights += np.dot(inputs.T, adjustments)
    
print("Weights after training")
print(weights)

print("Output after training")
print(activated_output)
```
